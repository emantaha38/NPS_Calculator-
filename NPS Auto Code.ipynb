```python
def nps (x):
    for i in str(x):
        if x > 8:
            return "Promoters"
        elif x <3:
            return "HD"
        elif x <7:
            return "Detractors"
        else:
            return "Neutral"
data['NPS score']=data['Q1'].apply(nps)
```

```python

# Define the calculate_metrics function
def calculate_metrics(data_nps):
    data_nps['Total_score'] = data_nps[['Detractors', 'HD', 'Neutral', 'Promoters']].sum(axis=1)
    
    data_nps['HD_percent'] = np.where(data_nps['Total_score'] == 0, 0, (data_nps['HD'] / data_nps['Total_score'] * 100).round().astype(int))
    data_nps['HD_percent'] = data_nps['HD_percent'].apply(lambda x: f"{x}%")
    data_nps['Promoters_Num'] = data_nps['Promoters'] - (data_nps['Detractors'] + data_nps['HD'])
    
    data_nps['Promoters_Percent'] = np.where(data_nps['Total_score'] == 0, 0, (data_nps['Promoters_Num'] /data_nps['Total_score'] * 100).round().astype(int))
    data_nps['Promoters_Percent'] = data_nps['Promoters_Percent'].apply(lambda x: f"{x}%")
    
    return data_nps

```

```python
result = calculate_metrics(data_nps.copy())
```
